```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\ABHISHEK YADAV\OneDrive\Desktop\crime_dataset_india.csv")
df
```

```python
# Q1. City-wise total crime count ka bar chart banao. Kaun sa city me sabse zyada crimes hue?
a=df['City'].value_counts()
plt.figure(figsize=(10,10))
plt.barh(a.index,a.values)
plt.xlabel('crime count')
plt.ylabel('City')
plt.title('City-wise total crime count ')
```

```python
# Q2. Crime Domain (Violent Crime vs Other Crime) ka bar chart banao.Kaunsa domain zyada common hai?

filter_data=df[df['Crime Domain'].isin(['Violent Crime','Other Crime'])]
a=filter_data['Crime Domain'].value_counts()
plt.figure(figsize=(10,5))
plt.bar(a.index, a.values)
plt.xlabel('Crime Domain')
plt.ylabel('Number of Crimes')
plt.title('Crime Domain Distribution')
for i,j in enumerate(a.values):
     plt.text(i,j,str(j),ha='center')
plt.show()




# -------------THIS IS FOR ALL CRIME DOMAIN DATA CRIME ----------------------
# df['Crime Domain'].value_counts()
# plt.figure(figsize=(20,10))
# plt.bar(a.index, a.values)

# plt.xlabel('Crime Domain')
# plt.ylabel('Number of Crimes')
# plt.title('Crime Domain Distribution')
# for i,j in enumerate(a.values):
#     plt.text(i,j,str(j),ha='center')

# plt.show()
```
