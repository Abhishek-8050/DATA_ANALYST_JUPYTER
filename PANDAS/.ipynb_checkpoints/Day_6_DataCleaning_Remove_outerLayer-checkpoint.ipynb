```python
import numpy as np
import matplotlib.pyplot as plt
a=np.array([20000,30000,50000,1500000,70000,90000])
```

```python
plt.boxplot(a)
```

```python
q1=np.quantile(a,0.25)
q2=np.quantile(a,0.50)
q3=np.quantile(a,0.75)
print(q1,q2,q3)
```

```python
IQR=q3-q1
IQR
```

```python
upper=q3+(1.5*IQR)
upper
```

```python
lower=q1-(1.5*IQR)
lower
```

```python
a
```

```python
v=[]
for i in a:
    if i>upper or i<lower:
        v.append(i)
print(v)
        
```

```python
import statistics as st
b=st.mean(v)
b
```
